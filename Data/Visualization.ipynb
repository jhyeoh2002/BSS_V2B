```python
import matplotlib.pyplot as plt
import pandas as pd
import os 
plt.rcParams["font.family"] = "Times New Roman"

BatteryDataFull = pd.read_csv('Battery_Data.csv')
BatteryData = pd.read_csv('Battery_Data_0129_0529.csv')

BatteryDataFull['datetime'] = pd.to_datetime(BatteryDataFull['datetime'])
BatteryDataFull.set_index('datetime', inplace=True)

BatteryData['datetime'] = pd.to_datetime(BatteryData['datetime'])
BatteryData.set_index('datetime', inplace=True)
```

```python
plt.figure(figsize=(12, 6), dpi=500)
plt.grid(True, alpha=0.3)
plt.title('BSS Data from Both Stations', fontsize=12)

# Mark missing data with red vertical lines, adding the label only once
missing_label_added = False
for i in range(len(BatteryData)):
    if pd.isna(BatteryData['stationA'][i]) or pd.isna(BatteryData['stationB'][i]):
        if not missing_label_added:
            plt.axvline(x=BatteryData.index[i], color='r', alpha=0.2, linewidth=1, label="Missing")
            missing_label_added = True
        else:
            plt.axvline(x=BatteryData.index[i], color='r', alpha=0.2, linewidth=1)

# Scatter plot for battery data from Station A and Station B
plt.scatter(BatteryData.index, BatteryData['stationA'], marker='|', linewidths=1.2, label="Station A")
plt.scatter(BatteryData.index, BatteryData['stationB'], marker='|', linewidths=1.2, label="Station B")

# Set axis labels and tick formatting
plt.ylabel('Number of Fully-charged Batteries, N', fontsize=10)
plt.xlabel('Date', fontsize=10)
plt.xticks(rotation=45)

# Adjust x-ticks to reduce overcrowding
plt.xticks(BatteryData.index[::168], rotation=45)  # Adjust interval as needed
plt.xlim([BatteryData.index[0],BatteryData.index[-1]])

# Add legend, improve layout, and save the figure
plt.legend(loc='best', fontsize=10)
plt.tight_layout()
plt.savefig('../Figures/BSS_Data_Availability.png', dpi='figure', format='png')
plt.show()
```
